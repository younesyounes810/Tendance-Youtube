## On enleve les signes de ponctuation et les caracteres speciaux 

```python
US_comments['comment_text'] = US_comments['comment_text'].str.replace("[^a-zA-Z#]", " ")
```

```python
US_comments['comment_text'] = US_comments['comment_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
US_comments['comment_text'] = US_comments['comment_text'].apply(lambda x:x.lower())
```

# Tokenization

```python
tokenized_tweet = US_comments['comment_text'].apply(lambda x: x.split())
tokenized_tweet.head()
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
wnl = WordNetLemmatizer()
```

```python
tokenized_tweet.apply(lambda x: [wnl.lemmatize(i) for i in x if i not in set(stopwords.words('english'))]) 
tokenized_tweet.head()
```

```python
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
```

## On aborde le Sentiment Analysis pour les commentaires de la Dataset

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

## On regle le "Sentiment Scores"

```python
US_comments['Sentiment Scores'] = US_comments['comment_text'].apply(lambda x:sia.polarity_scores(x)['compound'])
```

```python
US_comments.head()
```

## On classifie le " Sentiment scores" en Positif, negatif et Neutre

```python
US_comments['Sentiment'] = US_comments['Sentiment Scores'].apply(lambda s : 'Positive' if s > 0 else ('Neutral' if s == 0 else 'Negative'))
```

## On calcule le pourcentage de commentaires positifs dans toutes les vidéos.

```python
videos = []
for i in range(0,US_comments.video_id.nunique()):
    a = US_comments[(US_comments.video_id == US_comments.video_id.unique()[i]) & (US_comments.Sentiment == 'Positive')].count()[0]
    b = US_comments[US_comments.video_id == US_comments.video_id.unique()[i]]['Sentiment'].value_counts().sum()
    Percentage = (a/b)*100
    videos.append(round(Percentage,2))
```

## On Crée un dataframe des vidéos avec leurs pourcentages positifs.

```python
Positivity = pd.DataFrame(videos,US_comments.video_id.unique()).reset_index()
```

```python
Positivity.columns = ['video_id','Positive Percentage']
```

```python
Positivity.head()
```

##  On ajoute le nom de la chaîne des vidéos qui sont dans notre nouvel Dataset

```python
channels = []
for i in range(0,Positivity.video_id.nunique()):
    channels.append(US_videos[US_videos.video_id == Positivity.video_id.unique()[i]]['channel_title'].unique()[0])
```

```python
Positivity['Channel'] = channels
```

```python

tags = US_videos.tags.str.split('|').tolist()
# then get rid of anything that isn't a list
tags = [x for x in tags if type(x) == list]

# that gave us a list of lists (of strings), so we must separate the items in each 
tags2 = []
tags3 = []
for item in tags:
    for string in item:
        # get rid of numbers and other types
        if type(string) == str:
            tags2.append(string)

def meaningless(x):
    words = ['to','the','a','of','and','on','in','for','is','&','with','you','video']
    return x in words

# now let's split these strings by the spaces between words
for multiple in tags2:
    singles = multiple.split()
    # then let's add these cleaned tags to the final list
    for tag in singles:
        # now let's make everything lowercase and get rid of spaces
        tag = tag.strip()
        tag = tag.lower()
        # now let's remove the meaningless tags   
        if not meaningless(tag):
            tags3.append(tag)

# let's bring that into a dataframe
tagsdf = pd.DataFrame(tags3,columns=['tags'])
# then count the values
tagcounts = tagsdf.tags.value_counts()

# now preparing a bar chart representing the top values
tagcountslice = tagcounts[:30].sort_values()
tagcountslice.plot(kind='barh',title='Most Popular Tags in Trending Videos',grid=True,fontsize=12,figsize=(11,8))
plt.xlabel('In How Many Videos the Tag Occurred')

plt.tight_layout()
plt.show()
```
